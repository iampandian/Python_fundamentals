```python
#Dictionaries with one column of the data

country_code = {'India': 1, 'USA': 2, 'UK': 3, 'Japan': 4, 'China': 5}
country_code 
```

```python
type(country_code)
```

```python
#dict keys

country_code.keys()
```

```python
#dict values

country_code.values()
```

```python
#change value

country_code['China'] = 0
country_code
```

```python
del country_code[4]
country_code
```

```python
#contional loop, for 

for i in range(9):
    print(i)
```

```python
#while loop
a = 5
while (a <= 10):
    print(a)
    a = a+1
```

```python
#if condition
a = 1
if a ==1:
    print(a)
```

```python
#if else condition
a = 4
if a == 2:
    print(a)
else:
    print(' a is larger')
```

```python
#if elif else condition
a = 25
if a == 2:
    print('a is not 2')
elif a > 30:
    print("a is smaller than 30")
elif a % 5==0:
    print(a)
else:
    print('this is not a')
```

```python
for i in range(10):
    for j in range(10):
        if (i+j) % 2 == 0:
            print(f"the addition of {i} and {j} is divible by 2: {i+j}")
```

```python
#list comprehensive

[i for i in range(6)]
```

```python
# list comprehensive with condition

[i for i in range(6) if i%2 ==0]
```

```python
# Identify the number odd or even

[f'{i} is Even' if i%2 ==0 else f'{i} is Odd' for i in range(10)]
```
