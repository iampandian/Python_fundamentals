```python
# Now creating class and design a constructor method?

class Car:
    #constructor
    def __init__(self, brand, passengers_type, color, engine_type):
        self.brand = brand
        self.passengers_type = passengers_type
        self.color = color
        self.engine_type =engine_type
```

```python
car1 = Car('BMW', 4, 'Green', 'Petrol')

print(car1.__dict__)
```

Here, we saw two constructors, `__init__` is initialize the parameter(s) and `__dict__` is a dictionary format

```python
# inheritance 

class Car:
    #parent constructor
    def __init__(self, brand, door, color, engine_type, speed):
        self.brand = brand
        self.door = door
        self.color = color
        self.engine_type = engine_type
        self.speed = speed
        
    def car_speed(self):
        self.speed = self.speed + 5
        print(self.speed)
```

```python
car1 = Car('BMW', 4, 'Green', 'Petrol', 34)
car2 = Car('Audi', 4, 'Black', 'Petrol', 29)

car1.car_speed()
```

An instance is a specific object created from a particular class. To create instances of a class, call the class using the class name and pass in whatever arguments its `__init__` method acceptsâ€”in this example, the `__init__` method takes brand, door, color, engine_type and speed.
